```python
def fool_collect_transcript_links(driver, list_of_links=[], i=1, num_links=1000):
    earnings_calls_general = 'https://www.fool.com/earnings-call-transcripts'
    earnings_calls_page_extension = '/?page='
    while len(list_of_links) < num_links:
        url = earnings_calls_general + earnings_calls_page_extension + str(i)
        driver.get(url)
        page = driver.page_source
        for link in BeautifulSoup(page, parse_only=SoupStrainer('a')):
            if link.has_attr('href'):
                if re.match('/earnings/call-transcripts/.*', str(link['href'])):
                    list_of_links.append('https://www.fool.com'+link['href'])
        i+=1
        time.sleep(randrange(2,5))
        #Pdb().set_trace()
    return list_of_links, i
```

```python
def get_transcript_from_link(url):
    driver.get(url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    title = soup.find('h1').text
    call_transcript = soup.find(class_='article-content')
    try:
        call_transcript.find(text=re.compile('10 stocks we like better')).parent.parent.parent.parent.extract()
    except:
        print('Filed 10 stocks we like: '+title)
    try:
        ticker = [links.extract() for links in call_transcript.find_all('a', href=True)][0].text
    except:
        print('failed the links'+title)
    try:
        call_transcript.find(class_='image imgR').extract()
    except:
        print('failed remove image:'+title)
    return title, call_transcript


def get_all_transcripts(list_of_links, i=0, my_documents=[]):
    my_list = list(set(list_of_links))
    for x in range(i, len(my_list)):
        try:
            url = my_list[x]
            the_doc = get_transcript_from_link(url)
            print(the_doc[0])
            my_documents.append(the_doc)
            time.sleep(randrange(0,4))
        except:
            print('error error!')
            return my_documents
    return my_documents
#         if x%10 ==0 & x/10 >= 1:
#             file1 = open(file_name, "a")  # append mode
#             for y in range(0:10):
#                 file1.write(
#             file1.close()
        
```
