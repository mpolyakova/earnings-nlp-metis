```python
#What do I not want to do? 
#Lose part of speech tagging when I remove the stop words. 
#Can I preserve the tokens? 
#https://stackoverflow.com/questions/62760660/remove-stop-words-from-spacy-doc-object
#Boo, not the structure. 

def remove_punc_and_stop(text):
    clean_doc = []
    doc = nlp(text)
    for token in doc:
        if not token.is_stop and not token.is_punct and not token.like_num and token.text.isalpha():
            clean_doc.append(token)
    return clean_doc

def clean_general(text_list):
    nouns_root = [chunk.root for chunk in text_list.noun_chunks]
    adj = [ child.text for noun in nouns_root for child in noun.children if child.pos_ =='ADJ']
    nouns = [noun.text for noun in nouns_root]
    print(nouns+adj)
    for token in doc:
        if not  token.like_num and \
        (token.text in nouns+adj ) and token.text not in appends\
        :
            appends.append(token.lemma_)
    clean_doc.append(appends)


def print_test(list_of_tokens):
    for i in list_of_tokens:
        print(i.pos_)
        
        
text = 'And we project a return on beginning equity well above 20% in fiscal year 2022, driven by our permanent pivot to a more capital-efficient land strategy and our improved profitability.'
# a = remove_punc_and_stop(text)
# print(a)

# print(print_test(a))

# clean_general(a)
# #Perfect! The parts of speech are there 


# #But it appears the overall structure is lost - no more noun chunks. 
# #Does that mean noun chunks must be handled in the beginning - yes
# #https://stackoverflow.com/questions/62760660/remove-stop-words-from-spacy-doc-object


def create_doc(text):
    doc = nlp(text)
    return doc

def identify_ents(doc):
    #Identifies bigrams and groups. 
    #Since we're removing everything but noun and adjectives, we're fine losing the phrases attribute
    #for noun in list(doc.noun_chunks):
    #    noun.merge(noun.root.tag_, noun.root.lemma_, noun.root.ent_type_)
    #return doc 
    with doc.retokenize() as retokenizer:
        for ent in doc.ents:
            retokenizer.merge(ent)

    return doc



def identify_noun_chunks(doc):
#     #Identifies bigrams and groups. 
#     #Since we're removing everything but noun and adjectives, we're fine losing the phrases attribute
#     #for noun in list(doc.noun_chunks):
#     #    noun.merge(noun.root.tag_, noun.root.lemma_, noun.root.ent_type_)
#     #return doc 
    with doc.retokenize() as retokenizer:
        for noun in doc.noun_chunks:
               retokenizer.merge(noun)

    return doc

# def keep_nouns_and_adj(doc):
#     appends = []
#     nouns_root = [chunk.root for chunk in doc.noun_chunks]
#     adj = [ child.text for noun in nouns_root for child in noun.children if child.pos_ =='ADJ']
#     nouns = [noun.text for noun in nouns_root]
#     for token in doc:
#         if not token.like_num and \
#         (token.text in nouns+adj ) \
#         and token.text not in appends:
#             appends.append(token.text)
#     return appends

def keep_nouns_and_adj(doc):
    return_tokens=[]
    for token in doc:
        if token.pos_ in ['NOUN','ADJ']:
            return_tokens.append(token.text)
    return return_tokens
            
            
            
def remove_punct_and_stop(nouns_and_adj):
    token_list = identify_ents(nlp(nouns_and_adj))
    clean_list = []
    for token in token_list:
#         if token.ent_type in ['GPE','LANGUAGE','MONEY','NORP','ORG','PRODUCT','QUANTITY']:
#             print(token.text, '||', token.ent_type)
        if not token.is_stop \
        and not token.is_punct \
        and not token.like_num \
        and token.text.isalpha()\
        and token.ent_type_ not in('DATE','PERSON','PERCENT','TIME')\
        :
            clean_list.append(token.text)
    return clean_list

# def remove_punct_and_stop_and_verbs(tokens):
#     for token in nlp(tokens):
#         if not token.is_stop \
#         and not token.is_punct \
#         and not token.like_num \
#         and token.text.isalpha()\
#         and token.ent_type_ not in('DATE','PERSON','PERCENT','TIME'):
#         print(token)
#     return token

def spacy_cleaning(text):
    doc = create_doc(text)
    doc2 = identify_ents(doc)
    doc3 = identify_noun_chunks(doc2)
    a = keep_nouns_and_adj(doc3)
    b = remove_punct_and_stop(' '.join(a))
    return b

text = 'This loss reflects the impact of our seasonally slower revenue in Q4 and the increased selling and marketing costs as we close out our fiscal year.'
doc = create_doc(text)
print(doc)
doc2 = identify_ents(doc)
print(doc2)
doc3 = identify_noun_chunks(doc2)
print(doc3)
a = keep_nouns_and_adj(doc3)
# print(a)
b = remove_punct_and_stop(' '.join(a))
print(b)


#Ok. Now bigrams: 

#No noun chunks. Too many extra pieces. 


def create_bigrams(text_list):
    bigram = gensim.models.Phrases(text_list, min_count=10, threshold=10)
    bigram_mod = gensim.models.phrases.Phraser(bigram)
    return (bigram,bigram_mod)

def apply_bigrams(bigram_mod, text_list):
    return ([bigram_mod[x] for x in text_list])
    
def trigrams(text_list, bigram, bigram_mod):
    trigram = gensim.models.Phrases(bigram[text_list], min_count=3, threshold=10)
    trigram_mod = gensim.models.phrases.Phraser(trigram)
    
    return [trigram_mod[bigram_mod[x]] for x in text_list ] 
    
def remove_most_common(text_list, remove = .05):
    clean_doc = []
    all_words = ' '.join([' '.join(sentence) for sentence in text_list]).split(' ')
    len_to_remove = int(len(set(all_words))*remove)
    c = Counter(all_words)
    words_to_remove = [i[0] for i in c.most_common(len_to_remove)]
    
    for sentence in text_list:
        appends = []
        for word in sentence:
            if not word in words_to_remove:
                appends.append(word)   
        clean_doc.append(appends)
        
    return clean_doc
    
bigram, bigram_mod = create_bigrams([b])
y = apply_bigrams(bigram_mod,[b])
z = trigrams(y, bigram, bigram_mod)
zzzz = remove_most_common(z, .1)
    
print(zzzz)  

```
