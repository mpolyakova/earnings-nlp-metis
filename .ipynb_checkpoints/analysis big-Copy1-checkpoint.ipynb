```python
#https://huggingface.co/yiyanghkust/finbert-tone 

# 
finbert = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone',num_labels=3)
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')

berty = pipeline("sentiment-analysis", model=finbert, tokenizer=tokenizer)
```

```python
results = berty(list(prep.prepared_sentences))
# prep['label'] = [i['label'] for i in results]
# prep['score'] = [i['score'] for i in results]
# prep['subjectivity'] = [TextBlob(text).subjectivity for text in list(prep.prepared_sentences)]
```

```python
results = []
sentiment = []
print(len(list(prep.prepared_sentences)))
```

```python
array = list(prep.prepared_sentences)
```

```python
for i in range(len(results), len(array)):
    holder = berty([array[i]])[0]['label']
    sentiment.append(TextBlob(array[i]).subjectivity)
    #print(holder)
    results.append(holder)
```

```python
len(results)
len(sentiment)

file = open('sentiment_label.txt','w')
for doc in results:
    file.write('\n********************************************************************\n'+str(doc))
file.close()

file = open('subjectivity.txt','a')
for doc in sentiment:
    file.write('\n********************************************************************\n'+str(doc))
file.close()
```
